```python
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv("NEWS_API_KEY")
```

```python
newsapi = NewsApiClient(api_key=api_key)
```

```python
top_headlines = newsapi.get_top_headlines(q='bitcoin')
```
